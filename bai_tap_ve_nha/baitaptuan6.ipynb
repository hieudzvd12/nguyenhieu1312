 W6A4 
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp 
lại, thêm vào list của key đó

```python
n = input().split()
dick = {}
for m in n:
    key,value = m.split(":")
    if key not in dick:
        dick[key] = [value]
    else:
        old = dick[key]
        val_new = old.append(value)
        dick[key] = val_new
print(dick)
```

Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. 
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất

```python
dayn = list(map(int, input().split()))
dick = {}
for i in dayn:
    if i not in dick:
        dick[i] = 1
    else:
        dick[i] = dick[i] + 1
xuat_hien_nhieu = 0
so_lan_nhieu = ''
for i,xuat_hien in dick.items():
    if xuat_hien > xuat_hien_nhieu:
        xuat_hien_nhieu = xuat_hien
        so_lan_nhieu = i
    elif xuat_hien == xuat_hien_nhieu and i < so_lan_nhieu:
        so_lan_nhieu = i
print(so_lan_nhieu)

```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key
value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
ki = input().split()
k = int(input())
dick = {}
for i in range(0,len(ki),2):
    chia_khoa = ki[i]
    gia_tri = ki[i+1]
    if int(gia_tri) > k:
        dick[chia_khoa] = gia_tri

print(dick)


```
