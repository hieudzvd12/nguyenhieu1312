W5A3-Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def kiem_tra_sont(a):
    la_so_nt = True
    for i in range(2,a):
        if a % i == 0:
            la_so_nt = False
            break
    if la_so_nt:
        return "la so nguyen to"
    else:
        return "khong la so nguyen to"
a = int(input())
print(kiem_tra_sont(a))
```

W5A4-VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_hoan_hao(a):
    tong_cac_uoc = 0
    for i in range(1,a):
        if a % i == 0:
            tong_cac_uoc = tong_cac_uoc + i
    if tong_cac_uoc == a :
        return "la so hoan hao"
    else:
        return "khong la so hoan hao"
a = abs(int(input()))
print(so_hoan_hao(a))
```

W5A5-Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def gia_tri_va_day(list_a, k):
    for i in list_a:
        if i == k:
            return list_a.index(i)
    else:
        return "-1"
list_a = list(map(int, input().split()))
k = int(input())
print(gia_tri_va_day(list_a , k))
```

W5A6-Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def giai_thua(n):
    giai_thua_n = n
    for i in range(1,n):
        giai_thua_n = giai_thua_n*i
    return giai_thua_n
n = int(input())
print(giai_thua(n))
```

W5A8-Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 


```python
def khoang_cach_hamming(x,y):
    n = x^y
    dem = 0
    while n:
        n = n & (n - 1)
        dem = dem + 1
    return dem
x , y = map(int , input().split())
print(khoang_cach_hamming(x,y))

    
    
```

```python
def tong_chu_so(n):
    if n == 0:
        return "1"
    else:
        tong = 0
        while n > 0:
            tong = tong + 1
            n = n // 10
    return tong
n = int(input())
print(tong_chu_so(n))
```

W5A12-Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return "Khong the xay ra"
    else:
        q = dividend // divisor
        r = dividend % divisor
        return q , r
dividend , divisor = map(int , input().split())
print(divide_with_remainder(dividend,divisor))
    
```

W5A14-Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a , b , op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a*b
    elif op == "/":
        if b == 0:
            return "khong the xay ra"
        else:
            return a/b
a , b = map(int, input().split())
op = input()
print(compute(a , b , op))
    
```

W5A15-Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a , b , c):
    if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        return "True"
    else:
        return "False"
a , b ,c = map(int , input().split())
print(is_right_triangle(a , b , c))
```
