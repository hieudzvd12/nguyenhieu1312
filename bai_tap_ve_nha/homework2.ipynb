```python
##Bai5
def giai_pt_bac2(a , b ,c):
    delta = b**2 - 4*a*c
    if delta < 0:
        print("phuong trinh vo nghiem")
    elif delta == 0:
        x = (-b)/2*a
        print("phuong trinh co nghiem kep: ", x)
    elif delta > 0:
        x1 = (-b + delta**(1/2))/2*a
        x2 = (-b - delta**(1/2))/2*a
        print("phuong trinh co 2 nghiem phan biet: ",x1 , x2)
a , b , c = map(float , input().split())
giai_pt_bac2(a , b ,c)
```

```python
#Bai8
def giai_he_pt2(a , b , c , d , m , n):
    if a*d - b*c == 0 and m*d - b*n != 0 and a*n - m*c != 0:
        print("he vo nghiem")
    elif a*d - b*c == m*d - b*n == a*n - m*c == 0:
        print("he co vo so nghiem")
    elif a*d - b*c != 0:
        x = (m*d - b*n)/(a*d - b*c)
        y = (a*n - m*c)/(a*d - b*c)
        print("he co nghiem duy nhat: " , x,y)
a , b , c , d , m , n = map(float, input().split())
giai_he_pt2(a , b , c , d , m , n)
```
